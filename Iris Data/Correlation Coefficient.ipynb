## Purpose

In this study we will be using the famous Iris dataset from U.C. Irvine's machine learning repository website. This dataset contains four different measurements on the iris flowers of three different varieties. We will be investigating the correlations between these measurements.

## Step One: 

We need to load some necessary python libraries in order to perform the analysis.

```python
import os
import pandas as pd

import urllib
from urllib.request import urlretrieve
```

We use the following python command to print out the current working directory.

```python
print(os.getcwd())
```

## Step Two: 

Next we extract the Iris dataset directly from the UC Irvine machine learning repository and save it in our current working directory with the following python commands.

```python
if not os.path.exists('iris.txt'):
    urllib.request.urlretrieve('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',
                       'iris.txt')
```

Since the columns of the Iris dataset are not labeled, we create labels for these columns with the following python command.  

```python
iris_col = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
```

With the following python command we import the Iris dataset from our current working directory to this notebook and add our labels with the names option command in pd.read_table() function. 

```python
iris = pd.read_table('iris.txt', sep = ',', engine = 'python', header = None, names = iris_col)
```

We use the head function command to display the first five rows of the Iris dataset, as a check.  

```python
iris.head()
```

## Step Three: 

Now we examine the structure of the Iris dataset.

```python
iris.info(memory_usage = 'deep')
```

**How is the Iris dataset structured?**

1. This dataset is a pandas DataFrame.

2. It contains 150 entries from 0 to 149.

3. It has 5 columns: sepal_length, sepal_width, petal_length, petal_width, and varietal (class)

4. The first four of these columns are float type data; the last class is object type data.

5. Memory usage of this dataset is 15.1 KB.

**Attribute Information:**

 - The first four columns are measurements in centimeters.
 
 - Class refers to the varieties: 
 
     - Iris Setosa
     
     - Iris Versicolor 
     
     - Iris Virginica

```python
iris_setosa = iris[iris['class'] == 'Iris-setosa']
iris_versicolor = iris[iris['class'] == 'Iris-versicolor']
iris_virginica = iris[iris['class'] == 'Iris-virginica']
```

Loading the following two python libraries will enable amiable visualization of plots.

```python
%matplotlib inline

import seaborn as sns
```

## Step Four: 

The following python command will display a scatterplot matrix. 

```python
sns.pairplot(iris, vars = ['sepal_length', 'sepal_width', "petal_length", 'petal_width'],
                 hue = 'class', markers = '.');
```

**Multi-colored histograms** on the diagonal represent three levels of the target variable, class: **blue** for iris-setosa, **green** for iris-versicolor, and **red** for iris-virginica. The correlation scatterplots off the diagonal are mirror images across the diagonal with axes interchanged, and the same color convention.

- The **blue bars** on the histograms on the diagonal of the above scatterplot matrix are respectively the four measurements of the Iris setosa. The **green bars** are respectively the four measurements of the Iris versicolor. The **red bars** are respectively the four measurements of the Iris virginica.


- The **blue bars** in the histogram of the **sepal length** are clusttered to the left, indicating that the sepal length of the Iris setosa is generally **shorter** than the others; on the other hand, the **red bars** clustering to the right indicate that the sepal length of the Iris virginica is generally **longer** than the others; the **green bars** in the middle indicate that the sepal length of the Iris versicolor is generally midsize compared to the others.


- The **green bars** in the histogram of the **sepal width** are clustered to the left, indicating that the sepal width of the Iris versicolor is generally **narrower** than the others; on the other hand, the **blue bars** clustering to the right indicate that the sepal width of the Iris setosa is generally **wider** than the others; the **red bars** in the middle indicate that the sepal width of Iris virginica is generally midsize compared to the others.


- The **blue bars** in the histogram of the **petal length** are clustered to the left, indicating that the petal length of the Iris setosa is generally **shorter** than the others; on the other hand, the **red bars** clustering to the right indicate that the petal length of the Iris virginica is generally **longer** than the others; the **green bars** in the middle indicate that the petal length of the Iris versicolor is generally midsize compared to the others. 


- The **blue bars** in the histogram of the **petal width** are clustered to the left, indicating that the petal width of the Iris setosa is generally **shorter** than the others; on the other hand, the **red bars** clustering to the right indicate that the petal width of the Iris virginica is generally **longer** than the others; the **green bars** in the middle indicate that the petal width of the Iris versicolor is generally midsize compared to the others. 

The **second row and the first column** of the scatterplot matrix above shows a comparison of the sepal length and the sepal width of the three species of iris flowers. 

- The **blue dots** on this plot indicate that there is **strong positive** association between the sepal length and the sepal width of the Iris setosa. 


- The **red dots** on this plot indicate that there is **moderately positive** association between the sepal length and the sepal width of the Iris virginica.


- The **green dots** on this plot indicate that there is **moderately positive** association between the sepal length and the sepal width of the Iris versicolor.


- In **all three cases**, the slope of the Iris setosa (blue dots) is steeper than the other two Iris species; also, the blue dots are more compact than the other two groups. The scatterplot also shows that Iris setosa has shorter and wider sepals than the other two varietals.   

The **fourth row and the third column** of the scatterplot matrix above shows a comparison of the petal length and the petal width of the three species of iris flowers. 

- The **blue dots** on this plot indicate that there is **strong positive** association between the petal length and the petal width of the Iris setosa.


- The **green dots** on this plot indicate that there is **strong positive** association between the petal length and the petal width of the Iris versicolor.


- The **red dots** on this plot indicate that there is **less strong positive** association between the petal length and the petal width of the Iris virginica.


- In **all three cases**, petals of the Iris setosa are shorter and narrower than the other two varietals; also, the set of dots of the Iris setosa is much compact than other two groups. In fact, petals of the Iris virginica are longest and widest.    


- It is relatively pointless to examine the remaining four scatterplots, since they correlate a petal measurement with a sepal measurement.

# Step Five:

Values of the correlation coefﬁcient r lie between -1 and 1; r values closer to 0 indicate a weak relationship and r values closer to -1 or 1 indicate a strong relationship. Let's use the following rules to make our decisions.

```python
d = {'Positive' : pd.Series(['0.0 to 0.1', '0.1 to 0.3', '0.3 to 0.5', '0.5 to 1.0'], 
                            index = ['None', 'Weak', 'Moderate', 'Strong']),
     'Negative' : pd.Series(['0.0 to -0.1', '-0.1 to -0.3', '-0.3 to -0.5', '-0.5 to -1.0'], 
                            index = ['None', 'Weak', 'Moderate', 'Strong'])}
df = pd.DataFrame(d)
df.index.name = 'Strength of Association'
df
```

### Iris setosa:

The following python command produces the correlation coefficient table for the four measurements of the Iris setosa.

```python
iris_setosa.corr(method = 'pearson', min_periods = 1)
```

The values on the diagonal of the above correlation coefficient table are 1s, because the correlation of any field with itself is always perfect.


- The value of the correlation coefficient of the sepal width and sepal length is about 0.75, which says there is a strong positive correlation between the two fields. 


- The value of the correlation coefficient of the petal width and petal length is about 0.31, which says there is a low-to-moderate positive correlation between the two fields. 

### Iris versicolor:

The following python command produces the correlation coefficient table for the four measurements of the Iris versicolor.

```python
iris_versicolor.corr(method = 'pearson', min_periods = 1)
```

- The value of the correlation coefficient of the sepal width and sepal length of the Iris versicolor is about 0.53 which says there is a moderate-to-strong positive correlation between the two fields. 


- The value of the correlation coefficient of the petal width and petal length of the Iris versicolor is about 0.79 which says there is a strong positive correlation between the two fields. 

### Iris virginica:

The following python command produces the correlation coefficient table for the four measurements of the Iris virginica.

```python
iris_virginica.corr(method = 'pearson', min_periods = 1)
```

- The value of the correlation coefficient of the sepal width and sepal length of the Iris virginica is about 0.46 which says there is a moderate positive correlation between the two fields. 


- The value of the correlation coefficient of the petal width and petal length of the Iris virginica is about 0.32 which says there is a low-to-moderate positive correlation between the two fields. 


All these results are in agreement with the previous results from the scatterplot matrix.

As with the scatterplot matrix, it is relatively pointless to examine the remaining four correlations, since they involve a petal measurement with a sepal measurement.
